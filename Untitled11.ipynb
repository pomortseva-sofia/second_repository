```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('books.csv')
df1 = df[df['Год'] == 2023]
df2 = df[df['Год'] == 2024]

def line_chart(x, y):
  plt.subplot(2,1,1)
  plt.tight_layout()
  line = plt.plot(x, y, marker = 'o', color = 'red')
  plt.xlabel('Месяцы')
  plt.ylabel('Количество книг')
  plt.title('Прочитанные книги 2023')
  plt.legend(title = 'Прочитано книг')
  plt.grid(True)
  return line


def bar_chart(x, y):
  plt.subplot(2,1,2)
  plt.tight_layout()
  bar = plt.bar(x, y, color = 'green')
  plt.xlabel('Месяцы')
  plt.ylabel('Количество страниц')
  plt.title('Прочитанные страницы 2024')
  plt.grid(True)
  return bar


line_chart(df1['Месяц'], df1['Книги'])
bar_chart(df2['Месяц'], df2['Страницы'])
```
